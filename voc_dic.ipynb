```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.ver-taal.com/vocabulario.htm'
```

```python
# accessing the html content form webpage
response = requests.get(url)

# parsing the html content
soup = BeautifulSoup(response.content, 'html')

# get the title 
title = soup.title.text + '.csv'

title
```

```python
# extract the element from list that we will include it as header

headers_contenaire =[] # here we where store our hearders

headers = soup.select_one('ul', attri={'class':'nolist'})
type(headers)
```

```python
# Iterate all row in list in html
for row in headers:
    # romove the element which
    if row != '\n':
        headers_contenaire.append(row.text) # add each row to the list

print(headers_contenaire)
```

```python
# Remove Empty, underline elements
headers_contenaire = list(filter(None, headers_contenaire[0].split('\n')))

headers_contenaire
```

```python
# Now after we obtain the headers we will dive deeper into each element using it href to get it data

a = headers.find_all('a')
list_ = []
for href in a:
    array = []
    # Extract all links of rows
    url = 'https://www.ver-taal.com/' + href.get('href')
    print(url)

    # parsing each url using BeautifulSoup model
    url = requests.get(url)

    soup_2 = BeautifulSoup(url.content, 'html')

    table = soup_2.table

 
    for row in table.find_all('tr'):
        
        row = row.find('a')

        if row is not None:

            print(row.text)

            # add each collection to an array
            array.append(row.text)

            # Remove duplicate from array
            array = list(dict.fromkeys(array))

    # Store every collection in a contenaire     
    list_.append(array)

print(list_)


    
    
    
```
